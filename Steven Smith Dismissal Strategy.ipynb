#### Lets see in which areas does Steven Smith scores most runs in test cricket 

```python
runs_by_area = Steve_smith_test.groupby('Shot Area')['batruns'].sum().reset_index()

total_runs = runs_by_area['batruns'].sum()

runs_by_area['Percentage'] = round(((runs_by_area['batruns'] / total_runs) * 100),2)

runs_by_area = runs_by_area.sort_values(by='batruns', ascending=False)
runs_by_area
```

#### Lets see how does steven smith gets dismissed mostly in test cricket

```python
Steve_smith_dismissals = Steve_smith_test[Steve_smith_test['bat_out'] == True]

dismissal_counts = Steve_smith_dismissals.groupby('dismissal').size().reset_index(name='Count')

total_dismissals = dismissal_counts['Count'].sum()
dismissal_counts['Percentage'] = round(((dismissal_counts['Count'] / total_dismissals) * 100),2)

dismissal_counts = dismissal_counts.sort_values(by='Count', ascending=False)

print(dismissal_counts)
```

Similarly, when the ball is delivered on a similar line and length, around the 4th-5th stump line, but bounces up to chest height with some inswing, Smith is more likely to produce an inside edge.

So our Plan B is Divided into two parts :

1. Short of a Good Length, 4th-6th Stump Line, Bouncing to Stump Height (Stump Height Target)

 Objective: Induce Edges with Outswing

- Line: Focus on the 5th or 6th stump line, just outside off, encouraging Smith to defend the ball.
- Length: The ball should rise no higher than the stumps. This keeps Smith in his crease, making him vulnerable to swing.

 Variation: 
- Slight outswing is key. Balls that swing away from Smith at this height are difficult to leave, especially on the back foot.

 Dismissal Focus:
- Caught behind or in slips: Smith’s defensive technique can lead to an edge when the ball is angling away from him. The dismissals in 2023 and 2024 show that low-bouncing deliveries with outswing have consistently gotten Smith caught at point or in the slips.
- LBW: If the ball comes in slightly, Smith is prone to being trapped LBW. Inward movement can jam him as he’s stuck in his crease.


2. Short of a Good Length, 4th-5th Stump Line, Bouncing to Chest Height (Chest Height Target)

Objective: Use Extra Bounce to Pin Smith on the Back Foot

- Line: Target the 4th or 5th stump line. This is more aggressive, making Smith uncomfortable with the bounce.
- Length: The ball should rise to chest height. The steeper bounce forces Smith to play defensively, which is more likely to produce an inside edge.

Variation:
- Seam movement, especially if the ball jags in after pitching, creates significant discomfort for Smith. Chest-height deliveries challenge his balance and decision-making.

Dismissal Focus:
- Caught: Smith’s dismissals to Mark Wood and Mohammed Siraj were a result of short-of-good-length deliveries that climbed to chest height, forcing mistimed defensive shots that resulted in catches.
- Bowled: A ball that jagged back from the 4th-5th stump line has also led to Smith being bowled, as seen in 2023 and 2021.
